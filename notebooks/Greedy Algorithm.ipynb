N(1 ≤ N ≤ 100,000)개의 로프가 있다. 이 로프를 이용하여 이런 저런 물체를 들어올릴 수 있다. 각각의 로프는 그 굵기나 길이가 다르기 때문에 들 수 있는 물체의 중량이 서로 다를 수도 있다.

하지만 여러 개의 로프를 병렬로 연결하면 각각의 로프에 걸리는 중량을 나눌 수 있다. k개의 로프를 사용하여 중량이 w인 물체를 들어올릴 때, 각각의 로프에는 모두 고르게 w/k 만큼의 중량이 걸리게 된다.

각 로프들에 대한 정보가 주어졌을 때, 이 로프들을 이용하여 들어올릴 수 있는 물체의 최대 중량을 구해내는 프로그램을 작성하시오. 모든 로프를 사용해야 할 필요는 없으며, 임의로 몇 개의 로프를 골라서 사용해도 된다.

```python
N = int(input())
rope_limit = [int(input()) for _ in range(N)]
rope_limit = sorted(rope_limit, reverse=True)
# rope_limit = list(sorted([int(input()) for _ in range(N)], reverse=True))
```

```python
N = int(input())
rope_limit = [int(input()) for _ in range(N)]
rope_limit = sorted(rope_limit, reverse=True)

maximum_weight = max(rope_limit)
rope_cnt = 1
for rope in rope_limit:
    max_weigh = rope * rope_cnt
    if maximum_weight < max_weigh:
        maximum_weight = max_weigh
    rope_cnt += 1
print(maximum_weight)
```

# 1931. 회의실 배정

회의가 끝나는 순으로 정렬하여 푼다. 또한 sorted의 lambda의 인자를 2개 이상 주면 순서대로 적용한다.

```python
import sys
# input = sys.stdin.readline

N = int(input())

meet_time = sorted([tuple(map(int, input().split())) for _ in range(N)], key=lambda x: (x[1], x[0]))
meets = []

cur_end = 0
for start, end in meet_time:
    if start >= cur_end:
        cur_end = end
        meets.append(tuple([start, end]))
print(len(meets))
```

# 2579. 계단 오르기

```python
N = int(input())
stairs = [0] + [int(input()) for _ in range(N)]
```

```python
scores = [0 for _ in range(N+1)]
scores[1] = stairs[1]
scores[2] = stairs[1] + stairs[2]
```

```python
def climb(loc):
    if loc < 3:
        return scores[loc]
    
    for i in range(3, len(stairs)):
        # 직전 계단을 밟았을 경우, 다음 칸을 바로 못 밟아 2칸 떨어져야 하며,
        # 안 밟았으면 한 칸 이동이 가능하다.
        print(i)
        scores[i] = max(climb(i-2)+stairs[i], stairs[i-1] + stairs[i] + climb(i-3))
    
        
```
