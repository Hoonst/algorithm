# 20055. 컨베이어 벨트 위의 로봇

```python
list(range(10,-1,-1))
```

```python
N, K = map(int, input().split())
As = list(map(int, input().split()))

from collections import deque

turn = 1
As = deque(As)
robots = deque([0]*N)

while True:
    # 1. 벨트가 각 칸 위에 있는 로봇과 함께 한 칸 회전한다.
    As.rotate(1)  # 1 2 [1] > [1] 1 2
    
    # 한칸 회전함으로써 내리는 로봇에 대한 처리
    robots.rotate(1)
    robots[-1] = 0
    
    # 2. 가장 먼저 벨트에 올라간 로봇부터, 벨트가 회전하는 방향으로 한 칸 이동할 수 있다면 이동한다. 만약 이동할 수 없다면 가만히 있는다.
    if sum(robots):
        for i in range(N-2, -1, -1):
            # 로봇이 이동하기 위해서는 이동하려는 칸에 로봇이 없으며, 그 칸의 내구도가 1 이상 남아 있어야 한다.
            if robots[i] and robots[i+1] == 0 and As[i+1] > 0:
                robots[i+1] = 1 # move robot
                robots[i] = 0
                As[i+1] -= 1

        robots[-1] = 0
    
    # 3. 올리는 위치에 있는 칸의 내구도가 0이 아니면 올리는 위치에 로봇을 올린다.
    if robots[0] ==0 and As[0] > 0:
        robots[0] = 1
        As[0] -= 1
    
    # 4. 내구도가 0인 칸의 개수가 K개 이상이라면 과정을 종료한다. 그렇지 않다면 1번으로 돌아간다.
    zero_As = len([i for i in As if i == 0])
    if zero_As >= K:
        print(turn)
        break
    else:
        turn += 1
```

# 17140. 이차원 배열과 연산

```python
from collections import Counter

r, c, k = map(int, input().split())
graph = [list(map(int, input().split())) for _ in range(3)]

def function(array):
    for arr_idx in range(len(array)):
        counter = Counter(array[arr_idx])
        del counter[0]
        counter_tuple = [(i,counter[i]) for i in counter]

        to_list = sorted(counter_tuple, key=lambda x: (x[1], x[0]))
        to_list = [j for i in to_list for j in i]
        array[arr_idx] = to_list

    max_len = max([len(k) for k in array])
    
    for arr_idx in range(len(array)):
        diff_len = max_len - len(array[arr_idx])
        array[arr_idx] += [0] * diff_len
        array[arr_idx] = array[arr_idx][:100]
    array = array[:100]
        
    return array

def construct(array, r, c, k):
    turn = 0
    while turn < 101:
        if r-1 < len(array) and c-1 < len(array[0]) and array[r-1][c-1] == k:
            return turn
        else:
            # R 연산
            if len(array) >= len(array[0]):
                array = function(array)
#                 print('R Computation')
#                 print(array)

            # C 연산
            elif len(array) < len(array[0]):
                array = list(zip(*array))
                array = function(array)
                array = list(map(list, zip(*array)))
#                 print('C Computation')
#                 print(array)
            turn += 1
    return -1

print(construct(graph, r,c,k))
```

```python
from collections import deque
N, Q = map(int, input().split())
# range를 2**N으로 해야하는데 N으로 해서 틀렸지 뭐야?
graph = [list(map(int, input().split())) for _ in range(2**N)]
Ls = list(map(int, input().split()))

dr = [0,0,1,-1]
dc = [1,-1,0,0]

def rotate(input_grid, level):
    if level == 0:
        return input_grid
    
    cut_length = 2**level
    trial = len(graph) // cut_length
    
    new_grid = [[0]*(2**N) for _ in range(2**N)]
    for row in range(trial):
        for col in range(trial):
            grid = [row_cut[cut_length * (col):cut_length * (col + 1)] for row_cut in input_grid[cut_length * (row):cut_length * (row + 1)]]
            for i in range(len(grid)):
                one_row = grid[len(grid)-1-i]
                for element_idx in range(len(one_row)):
                    new_grid[(cut_length * row)+element_idx][(cut_length*col)+i] = one_row[element_idx]
                    
    return new_grid

def melting(input_grid):
    empty_list = []
    for row in range(len(input_grid)):
        for col in range(len(input_grid)):
            block = 0
            for i in range(4):
                if 0 <= row+dr[i] < len(input_grid) and 0 <= col+dc[i] < len(input_grid):
                    target = input_grid[row+dr[i]][col+dc[i]]
                    if target != 0:
                        block += 1
            if block < 3 and input_grid[row][col] > 0:
                empty_list.append((row, col))
    for r,c in empty_list:
        input_grid[r][c] -= 1
    return input_grid

def dfs(grid, r, c):
    cnt = 1
    queue = deque()
    queue.append((r,c))
    visited[r][c] = True
    while queue:
        r, c = queue.popleft()
        
        for i in range(4):
            nr = r + dr[i]
            nc = c + dc[i]
            if 0 <= nr < (2**N) and 0 <= nc < (2**N):
                if grid[nr][nc] > 0 and not visited[nr][nc]:
                    cnt += 1
                    visited[nr][nc] = True
                    queue.append((nr, nc))
    return cnt

for q in range(Q):
    L = Ls[q]
    rotated_graph = rotate(graph, L)
    graph = melting(rotated_graph)

max_size = 0
visited = [[False for _ in range(2**N)] for _ in range(2**N)]
total_sum = 0

for row in range(2**N):
    row_sum = sum(graph[row])
    total_sum += row_sum
    for col in range(2**N):
        if graph[row][col] and not visited[row][col]:
            _size = dfs(graph, row, col)
            if _size > max_size:
                max_size = _size

print(total_sum)
print(max_size)
```

```python
from copy import deepcopy

inp = '''
7 1
0 0 0 0 0 0 0
3 2 1 3 2 3 0
2 1 2 1 2 1 0
2 1 1 0 2 1 1
3 3 2 3 2 1 2
3 3 3 1 3 3 2
2 3 2 2 3 2 3
2 2
'''

inp = inp.strip().split('\n')
N, M = map(int, inp[0].split())
graph = [list(map(int, a.split())) for a in inp[1:N+1]]
moves = [list(map(int, a.split())) for a in inp[N+1:]]
moves = [[i-1, j] for i,j in moves]

# 파괴술식
def destruction(grid, spell):
    grid = deepcopy(grid)
    direction, power = spell
    destroy = [(center[0]+(dr[direction]*i), center[1]+(dc[direction]*i)) for i in range(1, power+1)]
    for row, col in destroy:
        grid[row][col] = 0
    return grid

def explode(list_box):
    index_1 = 0
    index_2 = 1
    
    one, two, three = 0,0,0
    
    flag = 0
    box_len = len(list_box)
    targets = []
    while list_box[index_1] > 0:
        first = list_box[index_1]
        length = 1

        while True:
            if first == list_box[index_2]:
                index_2 += 1
                length += 1
            else:
                break
        if length >= 4:
            flag = 1
            targets.append((index_1, index_2))
#             list_box = list_box[:index_1] + list_box[index_2:]
            index_1 = index_2
            index_2 = index_1 + 1
            
            if first == 1:
                one += length
            elif first == 2:
                two += length
            elif first == 3:
                three += length

        elif length < 4:
            index_1 += 1
            index_2 = index_1 + 1
    for t in targets:
        list_box[t[0]:t[1]] = [0] * (t[1]-t[0])
        
    list_box = [i for i in list_box if i != 0]
    list_box = list_box + [0] * (box_len - len(list_box))

    return list_box, flag, one, two, three

def transform(list_box):
    index_1 = 0
    index_2 = 0
    box_len = len(list_box)
    num_list = []
    
    while index_2 < len(list_box):
        first = list_box[index_1]
        length = 0
        while True:
            if index_2 == box_len:
                break
            
            if first == list_box[index_2]:
                index_2 += 1
                length += 1
            else:
                break
        
        num_list.append((length, first))
        index_1 = index_2
        
    num_list = num_list[:((N**2)-1)//2]
        
    return num_list

def recovery(input_box):
    new_grid = [[0 for j in range(N)] for _ in range(N)]
    
    hop = [i for i in range(1,N) for _ in range(2)] + [N-1]

    box_len = (N**2)-1
    center_r, center_c = center
    
    box_idx = 0
    for index, h in enumerate(hop):
        index %= 4
        for _ in range(h):
            center_r += roll_dr[index]
            center_c += roll_dc[index]
            new_grid[center_r][center_c] = input_box[box_idx]
            box_idx += 1
    return new_grid
```

중력이 작용하면 검은색 블록을 제외한 모든 블록이 행의 번호가 큰 칸으로 이동

```python
inp = '''
5 3
2 2 -1 3 1
3 3 2 0 -1
0 0 0 1 2
-1 3 1 3 2
0 3 2 2 1
'''

inp = inp.strip().split('\n')
N, M = map(int, inp[0].split())
graph = [list(map(int, a.split())) for a in inp[1:N+1]]


dr = [0,0,1,-1]
dc = [1,-1,0,0]
```

```python
from collections import deque

def bfs(grid, r, c, visited_):
    group = []
    queue = deque()
    queue.append((r,c))
    group.append((r,c))
    criterion = grid[r][c]
 
    while queue:
        r, c = queue.popleft()
        visited_[r][c] = True
        for i in range(4):
            nr = r + dr[i]
            nc = c + dc[i]
            
            if 0 <= nr < N and 0 <= nc < N and not visited_[nr][nc]:
                if grid[nr][nc] == 0:
                    if (nr,nc) not in queue and (nr,nc) not in group:
                        queue.append((nr, nc))
                        group.append((nr, nc))
                elif grid[nr][nc] == criterion:
                    visited_[nr][nc] = True
                    queue.append((nr, nc))
                    group.append((nr, nc))                    
    
    return group

# 1번 규칙: 가장 큰 블록 그룹 수색
def search_block(total_map):
    total_group = []
    new_group = []
    maxi = 0
    visited=[[False for _ in range(N)] for _ in range(N)]
    for r in range(N):
        for c in range(N):
            if visited[r][c] == False and total_map[r][c] > 0:
                found_group = bfs(total_map, r,c, visited)
                if len(found_group) >= 2:
                    total_group.append(found_group)
                    
    if total_group:   
        max_len = len(max(total_group, key=lambda x: len(x)))
        total_group = [i for i in total_group if len(i) == max_len]
        
        for idx, group in enumerate(total_group):
            group = sorted(group, key=lambda x: [x[0], x[1]])
            total_group[idx] = group
            
        total_group = min(total_group, key=lambda x: [x[0][0], x[0][1]])
    
    return total_group

def gravity(grid):
    for r in range(N-2, -1, -1):
        for c in range(N):
            if grid[r][c] > -1:
                temp = r
                while True:
                    if 0 <= temp+1 < N and grid[temp+1][c] == -2:
                        grid[temp+1][c] = grid[temp][c]
                        grid[temp][c] = -2
                        temp += 1
                    else:
                        break     
    return grid

def rotate(grid):
    new_grid = [[0]*N for _ in range(N)]
    for r in range(N):
        for c in range(N):
            new_grid[N-1-c][r] = grid[r][c]
    return new_grid

def erase(grid, block):
    for r,c in block:
        grid[r][c] = -2
    return grid
```

```python
point = 0
while True:
    searched = search_block(graph)
    print(f'searched: {searched}')
    if searched:
        point += len(searched) ** 2
        erased_graph = erase(graph, searched)
        gravity_on = gravity(erased_graph)
        rotated = rotate(gravity_on)
        
        graph = gravity(rotated)
        for g in graph:
            print(g)
        print(f'point: {point}')
    else:
        break
```
